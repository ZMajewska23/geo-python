```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plot
import seaborn as sns
import statsmodels as sm
```

```python
df_month_r = pd.read_csv('data/df_minutes_r.csv', sep=';', decimal=',')
df_month_k = pd.read_csv('data/df_minutes_k.csv', sep=';', decimal=',')

```

```python
df_month_r.insert(0,"Miasto", "Rybnik")
df_month_r


```

```python
df_month_k.insert(0,"Miasto", "Katowice")
df_month_k
```

```python
df_concat = pd.concat([df_month_r, df_month_k]).dropna()
df_concat
```

```python
df_concat = df_concat[(df_concat['przyspieszenie']) < 5]
```

```python
sns.set()
```

```python
g = sns.pairplot(data=df_concat[['przyspieszenie', 'cisnienie[mBar]', 'Miasto']], hue='Miasto', diag_kind = 'hist') #height=3, aspect=2, palette=('cubehelix'), markers='o')
```
